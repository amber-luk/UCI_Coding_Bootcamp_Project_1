```python
# Import Dependencies
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Import our data into pandas from CSV
hate_crime = 'hate_crime_copy.csv'
hate_crime_df = pd.read_csv(hate_crime, low_memory=False)
```

```python
#Check count of each column
col = hate_crime_df.columns
for x in col:
    state_group = hate_crime_df.groupby(x)
    count_states = state_group[x].count()
#     print(count_states)
```

```python
#Incidents Per Year
fig = plt.figure(figsize=(20, 5))
state_group = hate_crime_df.groupby(hate_crime_df['DATA_YEAR'])
count_states = state_group['DATA_YEAR'].count()
x_axis = np.arange(len(count_states))
plt.bar(x_axis, count_states, color='red', alpha=0.5, align="center")
years = np.unique(hate_crime_df['DATA_YEAR'])
tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, years,rotation = 90,fontsize=20)
plt.title("Incidents Per Year",fontsize=20)
plt.xlabel("Years",fontsize=20)
plt.ylabel("Incidents",fontsize=20)
plt.savefig('IncidentsPerYear.png', bbox_inches = 'tight')
```

```python
#Incidents By Division
fig = plt.figure(figsize=(20, 5))
state_group = hate_crime_df.groupby(hate_crime_df['DIVISION_NAME'])
count_states = state_group['DIVISION_NAME'].count()
count_states = count_states.sort_values(ascending=False)
x_axis = np.arange(len(count_states))
count_states.plot(kind='bar',fontsize=20)
plt.title("Incidents By Division",fontsize=20)
plt.xlabel("Division",fontsize=20)
plt.ylabel("Incidents",fontsize=20)
# plt.savefig('IncidentsByDivision.png', bbox_inches = 'tight')
```

```python
#Incidents by State
fig = plt.figure(figsize=(20, 5))
state_group = hate_crime_df.groupby(hate_crime_df['STATE_NAME'])
count_states = state_group['STATE_NAME'].count()
count_states.plot(kind='bar',fontsize=18)
plt.title("Incidents By State",fontsize=20)
plt.xlabel("State",fontsize=20)
plt.ylabel("Incidents",fontsize=20)
plt.savefig('IncidentsByState.png', bbox_inches = 'tight')
```

```python
#Incidents By Region
fig = plt.figure(figsize=(20, 5))
state_group = hate_crime_df.groupby(hate_crime_df['REGION_NAME'])
count_states = state_group['REGION_NAME'].count()
count_states = count_states.sort_values(ascending=False)
count_states.plot(kind='bar',fontsize=20)
plt.title("Incidents By Region",fontsize=20)
plt.xlabel("Region",fontsize=20)
plt.ylabel("Incidents",fontsize=20)
# plt.savefig('IncidentsByRegion.png', bbox_inches = 'tight')
```

```python
#Incidents By Offense
fig = plt.figure(figsize=(20, 5))
state_group = hate_crime_df.groupby(hate_crime_df['OFFENSE_NAME'])
count_states = state_group['OFFENSE_NAME'].count()
count_states = count_states.sort_values(ascending=False)[:10]
count_states.plot(kind='bar',fontsize=15)
plt.title("Incidents By Offense (Top 10)",fontsize=20)
plt.xlabel("Offense",fontsize=20)
plt.ylabel("Incidents",fontsize=20)
# plt.savefig('IncidentsByOff.png', bbox_inches = 'tight')
```

```python
#Group by size of city and offense name
size1_count = size1.groupby(["OFFENSE_NAME"]).count()["INCIDENT_ID"]
size2_count = size2.groupby(["OFFENSE_NAME"]).count()["INCIDENT_ID"]
size3_count = size3.groupby(["OFFENSE_NAME"]).count()["INCIDENT_ID"]
size4_count = size4.groupby(["OFFENSE_NAME"]).count()["INCIDENT_ID"]
size5_count = size5.groupby(["OFFENSE_NAME"]).count()["INCIDENT_ID"]
size6_count = size6.groupby(["OFFENSE_NAME"]).count()["INCIDENT_ID"]
size7_count = size7.groupby(["OFFENSE_NAME"]).count()["INCIDENT_ID"]
size8_count = size8.groupby(["OFFENSE_NAME"]).count()["INCIDENT_ID"]

#Combine groupby's to one dataframe
new = pd.concat([size1_count,size2_count,size3_count,size4_count,size5_count,size6_count,size7_count,size8_count],axis=1)

#Add Column names
symbols=['Cities 1,000,000+','Cities 500,000-999,999','Cities 250,000-499,999','Cities 100,000-249,000','Cities 50,000-99,000','Cities 25,000-49,999','Cities 2,500-9,999','Cities < 2,500']
new.columns=symbols

#Clean up data by adding Total
new['Total']=new[list(new.columns)].sum(axis=1)

#Sort by values (DOESN'T WORK)
new.sort_values('Total',ascending=False)

#Drop NAs
new = new.dropna()

#Drop anything less than 400 count
new = new[new.Total > 400]

#Drop Total's column
new=new.drop(columns=['Total'])

#Cleaned Up DataFrame
new.head(2)
```

```python
#Create Scatterplot
fig = plt.figure(figsize=(30, 10))
lst = list(new.index)
for crime in lst:
    mark=["s","v","<","^",">","o","p","h","+","x"]
    y = new.loc[crime].index
    x = new.loc[crime].values
    plt.scatter(x, y, label=crime,alpha=0.9,s=180,linewidths=1,marker=mark[lst.index(crime)])
plt.xticks(rotation='vertical')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5),fontsize=20)
plt.title("Offense Count by Size of City",fontsize=20)
plt.ylabel("Size of City",fontsize=20)
plt.xlabel("Incidents",fontsize=20)
plt.savefig('OffenseByCitySize.png', bbox_inches = 'tight')
```
